```python
import random

def functions(x: float, i: int):
    if i == 0:
        return x**2+x+12, 2*x+1, 11.75
    
    return x**4+3*(x**3)+17, 4*(x**3)+9*(x**2), 8.46

def gradient_decent(function_index, lr, optim):
    x = random.randint(-2, 0)
    while True:
        val, grad, min = functions(x, function_index)
        x = x - grad*lr
        print("x: ", x, ", y: ", val)
        if abs(val-min) <= optim:
            break

def gradient_decent_reverse(function_index, lr, optim):
    x = random.randint(-2, 0)
    while True:
        val, grad, min = functions(x, function_index)
        x = x + grad*lr
        print("x: ", x, ", y: ", val)
        if abs(val-min) <= optim:
            break

gradient_decent(2, 0.00001, 0.001)

'''
for many complicated functions, finding min + max is difficult, maybe impossible => need grad

'''
```
